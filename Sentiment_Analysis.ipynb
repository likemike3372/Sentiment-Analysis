```python
import numpy as np
from keras.datasets import imdb
from keras.models import Sequential
from keras.layers import Dense, LSTM
from keras.layers.embeddings import Embedding
from keras.preprocessing import sequence

np.random.seed(10) # seed fixes the randomness and generates the same number again and again
```

```python
top_words = 5000
imdb.load_data(num_words = top_words) ## considers only 5000 words and replaces the non top_words with oov_char aka Null
```

```python
(x_train,y_train),(x_test, y_test) = imdb.load_data(num_words=top_words)
```

```python
max_review_length = 500
X_train = sequence.pad_sequences(x_train, maxlen = max_review_length)
X_test = sequence.pad_sequences(x_test, maxlen = max_review_length)
```

```python
print(X_train.shape)
```

```python
print(X_train[0])
```

```python
y_train.shape
print(y_train[:10])
```

```python
embedding_vector_length = 32
model = Sequential()
model.add(Embedding(top_words,
                    embedding_vector_length,
                    input_length = max_review_length))

model.add(LSTM(100))

model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', ## similar to SGD
              optimizer='adam',
              metrics=['accuracy'])

model.summary()


## it learns the vector during backprop = this implies that the words keep on moving in the multidimensional plot
```

```python
model.fit(X_train,y_train,epochs=3, batch_size=64)
scores = model.evaluate(X_test,y_test, verbose=0) #avoiding lot of info verbose = 0

print("Accuracy {}".format(scores[1]*100)) ## using format string
```
